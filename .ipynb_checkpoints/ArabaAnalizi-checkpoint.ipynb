```python
dataFrame=yuzdeDoksanDokuzlukDf #artık 0.01 lik verileri attığımız veride çalışıyoruz
```

```python
dataFrame=dataFrame[dataFrame.year != 1970] #1970 yıllarını da attıp güncelledik
```

```python
dataFrame.groupby("year").mean()["price"]
```

```python
dataFrame.head()
```

```python
dataFrame=dataFrame.drop("transmission",axis=1)
```

```python
#artık model oluşturabilecek şekilde veri ön işlemini tamamladık ve model oluşturuyoruz.
```

```python
y=dataFrame["price"].values #.values ile bir numpy dizisine çevirdik
x=dataFrame.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,  x_test, y_train, y_test =train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler() #obje olusturduk
#scaler kısmı sadece x e uygulanır
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #model oluşturuyoruz
from tensorflow.keras.layers import Dense #katmanları oluşturuyoruz
```

```python
x_train.shape #katmanlarımızı besten başlatmak mantıklı daha büyük bir sayıda kullanabilriz
```

```python
model = Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi=pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
tahminDizisi=model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
